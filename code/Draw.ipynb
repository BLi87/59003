```python
from matplotlib import pyplot as plt 
  
# x-axis values 
x = [4, 10, 20, 30, 40, 50, 60] 
  
# Y-axis values 
y = [510,1071,1,1538,1,1,3048] 
  
# Function to plot 
plt.plot(x, y) 

# Adding Title
plt.title("Data Size: 1372")
  
# Labeling the axes
plt.xlabel("Dimension")
plt.ylabel("Convergence Time/s")
  
# function to show the plot 
plt.show()
```
